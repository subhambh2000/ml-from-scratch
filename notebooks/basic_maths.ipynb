# Statistical calculations and matrix operations

```python
# Input for statistical operations
data = np.array([1, 2, 3, 4])

x = np.array([1, 2, 3])
y = np.array([2, 4, 6])

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

```

```python
# Mean and Variance
mean = mean(data)
variance = variance(data)

print(f"Mean: {mean}")
print(f"Variance: {variance}")

np_mean = np.mean(data)
print(f"NP mean: {np_mean}")
np_var = np.var(data)
print(f"NP variance: {np_var}")

print(f"Difference in mean: {mean - np_mean}")
print(f"Difference in variance: {variance - np_var}")
```

```python
# Matrix transpose

original_matrix = [[1, 2, 3], [4, 5, 6]]
transposed_matrix = transpose(original_matrix)
np_transposed_matrix = np.transpose(original_matrix)

print_ndarray(original_matrix, label="Original Matrix")
print_ndarray(transposed_matrix, label="Transposed matrix")
print_ndarray(np_transposed_matrix, label="NP Transposed matrix")

```

```python
# Covariance

covariance_matrix = np.cov(x, y)
print_ndarray(covariance_matrix, label="covariance")

# Correlation
print_ndarray(np.corrcoef(x, y), label="correlation")
```

```python
# Normalization
normalized_data = (data - np.min(data)) / (np.ptp(data))
print_ndarray(normalized_data, label="normalized data")

# Standardization
standardized_data = standardize(data)
np_standardized_data = (data - np.mean(data)) / np.std(data)

print_ndarray(standardized_data, label="standardized data")
print_ndarray(np_standardized_data, label="standardized data")

data_1 = np.array([1, 1, 1, 1])
standardized_data_edge = standardize(data_1)
# np_standardized_data_edge = (data_1 - np.mean(data_1)) / np.std(data_1)

print_ndarray(standardized_data_edge, label="standardized data edge case")
# print_ndarray(np_standardized_data_edge, label="np standardized data edge case")
```

```python
# Mean squared error
y_true = np.array([3.0, -0.5, 2.0, 7.0])
y_pred = np.array([2.5, 0.0, 2.0, 8.0])

mse = mean_squared_error(y_true, y_pred)
np_mse = np.mean((y_true - y_pred) ** 2)

print(f"Mean Squared Error: {mse}")
print(f"NP Mean Squared Error: {np_mse}")
```

```python
# Dot Product and matrix multiplication

# Dot product
v1, v2 = a[0], b[0]
dot_product = dot_product(v1, v2)
np_dot_product = np.dot(v1, v2)
print(f"dot product: {dot_product}")
print(f"np dot product: {np_dot_product}")

# Matrix multiplication
matrix_product = a @ b
print_ndarray(matrix_product, label="matrix product")
```

# Calculus

## Differentiation (derivative)

Derivative: $f'(x) = \frac{d(f(x))}{dx}$

<br>Example: $f'(x^2 + \sin x) = \frac{d(x^2 + \sin x)}{dx}=2x+\cos (x)$

```python
x = sp.symbols('x')
f = x ** 2 + sp.sin(x)
# diff(expression, variable)
derivative = sp.diff(f, x)
print(f"derivative of x^2 sin(x) = {derivative}")
```

## Integration

$\int f(x)dx = F(x) + c$

<br>Example: $\int (2x + \cos x) dx = x^2 + \sin x + c$

```python
indefinite_integral = sp.integrate(sp.cos(x), x)
definite_integral = sp.integrate(x ** 2, (x, 0, 1))

print(f"Indefinite integral: {indefinite_integral}")
print(f"Definite integral from 0 to 1: {definite_integral}")
```

## Limits
$\lim_{x \to a} f(x) = L$

<br>Example: $\lim_{x \to 0} \frac{\sin x}{x} = 1$

```python
lim = sp.limit(sp.sin(x) / x, x, 0)
print(f"Limit as x approaches 0 of sin(x)/x: {lim}")
```
